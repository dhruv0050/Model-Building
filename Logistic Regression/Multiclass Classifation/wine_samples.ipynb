```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
```

```python
dir(wine)
```

```python
import matplotlib.pyplot as plt

_, ax = plt.subplots()
scatter = ax.scatter(wine.data[:, 0], wine.data[:, 1], c=wine.target)
ax.set(xlabel=wine.feature_names[0], ylabel=wine.feature_names[1])
handles, _ = scatter.legend_elements()
ax.legend(handles, wine.target_names, loc="lower right", title="Classes")
```

```python
wine.feature_names
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
model = LogisticRegression()
X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size=0.2, random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
sample1 = [[13.2, 2.77, 2.51, 18.5, 98.0, 2.4, 2.3, 0.26, 1.56, 5.68, 1.03, 3.17, 1185.0]]
pred1 = model.predict(sample1)
print("Predicted wine class name:", wine.target_names[pred1])
```

```python
sample2 = [[13.71, 5.65, 2.45, 20.5, 95.0, 1.68, 0.61, 0.52, 1.06, 7.7, 0.64, 1.74, 740.0]]
pred2 = model.predict(sample2)
print("Predicted wine class name:", wine.target_names[pred2])
```

```python
model.score(X_test, y_test)  # Evaluate the model on the test set
```
