The task is to predict per-capita income for the years there in Years.csv


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df.head(4)
```

```python
plt.scatter(df.year,df.income, color='red', marker='x')
plt.xlabel("Year")
plt.ylabel("Per Capita Income(USD)")
plt.title("Year vs Per Capita Income")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df.income)
```

```python
d = pd.read_csv('years.csv')
p = reg.predict(d)
d['income'] = p
d.to_csv('predicted_income.csv', index=False)
```

```python
d.head(4)
```

```python
plt.scatter(df.year,df.income, color='red', marker='x')
plt.xlabel("Year")
plt.ylabel("Per Capita Income(USD)")
plt.title("Year vs Per Capita Income")
plt.plot(df.year,reg.predict(df[['year']]))
```

Saving the model using pickle and using it

```python
# Saving the model
import pickle
with open('model_pickle','wb') as f:
    pickle.dump(reg, f)
```

```python
# Loading the model
with open('model_pickle', 'rb') as f:
    model = pickle.load(f)
```
