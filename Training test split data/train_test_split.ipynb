We have a csv file that contains information of students that shows the relation between their studying hours and marks.
We need to train a model and test it

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('study_scores.csv')
df.head()
```

```python
plt.scatter(df['Hours Studied'], df['Score'])
```

We can use linear regression for this

```python
X = df[['Hours Studied']]
y = df['Score']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 10)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test, y_test)  # To evaluate the model's performance
```
